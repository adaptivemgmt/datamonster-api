
## `Datasource` splits methods

```python
# Datasource APIs
a_1010data_source = next(dm.get_datasources(query='1010data Credit Sales Index'))
_1010_summary = a_1010data_source.get_splits_summary()
_1010_summary
```

### Summary & splits for `'1010data Credit Sales Index'`, company = THE GAP

```python
the_gap = dm.get_company_by_ticker('GPS')
_1010_gap_summary = a_1010data_source.get_splits_summary(company=the_gap)
_1010_gap_summary
```

```python
_1010_gap_splits = a_1010data_source.get_splits(company=the_gap)
_1010_gap_splits
```

### Summary & splits for `'1010data Credit Sales Index'`, company = Pier 1 Imports (PIR)

```python
# Another company for this datasource: Pier 1 Imports (ticker: PIR)
pier1 = dm.get_company_by_ticker('PIR')
pier1
```

```python
_1010_pier1_splits = a_1010data_source.get_splits(company=pier1)
_1010_pier1_splits
```

### Summary & splits for `'1010data Credit Sales Index'`, companies = [The Gap, Pier 1 Imports]
There are 5 = 4 + 1 splits:

```python
_1010_gps_pier1_summary = a_1010data_source.get_splits_summary(
                                                company=[the_gap, pier1])
_1010_gps_pier1_summary
```

### Other multiple-valued keywords

Get the splits for `'Banana Republic'` and then for both `'Banana Republic'` and `'Old Navy'`, among the splits of The Gap:

#### Just `'Banana Republic'`

```python
# Drill down to one of the four Gap splits, 'Banana Republic'
_1010_gap_br_splits = a_1010data_source.get_splits(company=the_gap,
                                                   category='Banana Republic')
# In fact, the `company` parameter is unnecessary:
_1010_br_splits = a_1010data_source.get_splits(category='Banana Republic')
assert _1010_gap_br_splits == _1010_br_splits

_1010_gap_br_splits
```

#### Both `'Banana Republic'` and `'Old Navy'`

```python
_1010_gap_br_navy_splits = a_1010data_source.get_splits(
                                                category=('Banana Republic', 'Old Navy'))
_1010_gap_br_navy_splits
```
